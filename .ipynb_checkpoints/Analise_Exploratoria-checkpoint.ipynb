# Músicas do Spotify - Análise Exploratória

```python
# Importando as bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Carregando os dados
songs = pd.read_csv("data/dataset.csv", index_col=0)
songs.head()
```

```python
# Checando os tipos de dados
songs.info()
```

## Limpeza e tratamento

```python
# Verificando valores ausentes
songs.isna().sum()
```

```python
# Verificamos as linhas com valores ausentes
songs[songs.isna().any(axis=1)]
```

Temos apenas um registro com os valores ausentes. Podemos removê-lo sem prejuízos posteriores

```python
# Removendo o registro 65900
songs = songs.drop(65900)
songs.isna().sum()
```

## Análises

### Algumas distribuições de frequência

```python
# Lista de características quantitativas
numFeatures = ["duration_ms", "danceability", "energy", "loudness", 
               "speechiness", "acousticness", "instrumentalness",
               "liveness", "valence", "tempo"]

# Lista de características qualitativas
catFeatures = [col for col in songs.columns if col not in numFeatures + ["track_id", "popularity"]]
```

```python
songs[numFeatures].describe()
```

```python
songs[numFeatures].hist(figsize=(10.5, 10));
```

**Observações**:
1. Interessante notar que `energy` segue uma distribuição crescente
2. `tempo`, `danceability` e `valence` aparentemente se aproximam de uma Gaussiana

### Relações entre variáveis numéricas

```python
corr = songs[numFeatures + ["popularity"]].corr()
```

```python
# Veremos as correlações entre as características musicais numericas
sns.heatmap(corr, annot=True, fmt=".2f")
plt.title("Correlação entre as características musicais", fontsize="x-large", alpha=.8);
```

**Observações**:
1. Aparentemente, nenhuma das variáveis contínuas possui alta correlação com `popularity` (nossa variável alvo)
2. `energy` e `acousticness` possuem alta correlação negativa, o que pode causar algum problema de multicolinearidade posteriormente, assim como `loudness` e `energy`.

### Análises nas Variáveis Qualitativas

```python
keys = ["C", "C#", "D", "D#", "E", "F", "F#", "G", "G#", "A", "A#", "B"]
for i in range(len(keys)):
    songs.loc[songs.key == i, "key"] = keys[i]

sns.countplot(songs, x="key")
plt.title("Musicas por tom");
```

**Observações**:
1. Temos poucas músicas na tonalidade de D#/Eb (Ré sustenido / Mi bemol)
2. A maior parte das músicas estão em C (Dó) ou G (Sol)

```python
sns.barplot(songs, x="key", y="popularity")
plt.title("Popularidade por Tom");
```

**Observações**:
1. As tonalidades possuem popularidade balanceada

```python
sns.countplot(songs, x="time_signature")
plt.title("Musicas por marcação de Tempo");
```

```python
sns.barplot(songs, x="time_signature", y="popularity")
plt.title("Popularidade por Tempo");
```

## Pre-processamento nos dados

```python
# Removemos atributos passíveis de multicolinearidade
songsClean = songs.copy()
songsClean = songsClean.drop(["acousticness", "loudness"], axis=1)

numFeatures = ["duration_ms", "danceability", "energy", "speechiness", 
               "instrumentalness", "liveness", "valence", "tempo"]
```

### Normalização dos dados quantitativos

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(songsClean[numFeatures])

songsClean[numFeatures] = scaler.transform(songsClean[numFeatures])
```

```python
songsClean.head()
```

### Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
for i in catFeatures:
    encoder = LabelEncoder()
    encoder.fit(songsClean[i])
    songsClean[i] = encoder.transform(songsClean[i])
```

```python
songsClean.head()
```

## Análise de atributos relevantes

```python
from statsmodels.api import OLS
```

```python
model = OLS(songsClean["popularity"], songsClean[numFeatures + catFeatures])
res = model.fit()

res.summary()
```

**Observações**:
1. Percebemos que o valor de R-Squared está relativamente expressivo
2. album_name, artists e track_name aparentemente são completamente irrelevantes para a popularidade

```python
# Iremos exportar o dataframe limpo para realizar a análise preditiva
songsClean.to_csv("data/songsClean.csv", index=False)
```
